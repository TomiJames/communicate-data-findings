## Investigation Overview


In this analysis, I explored the Ford GoBike system to understand what factors influence the duration of the trip. I primarily looked at the distance covered, the time of the day the trip was taken as well as the age and gender of the riders.


## Dataset Overview
The original dataset contained about 183,412 observations across 16 variables. Most of the variables speak to details about the trip itself and a few such as age and gender describe the rider. In the course of the analysis, new features were derived from existing ones. Also, outliers from the trip duration were also dropped.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('edited_201902-fordgobike-tripdata.csv')
```

```python
# change data type of time_of_day column to category
time_order = ['early morning','morning','mid-day','evening','night','late night']
ordered_time = pd.api.types.CategoricalDtype(ordered = True,categories = time_order)
df.time_of_day = df.time_of_day.astype(ordered_time)
```

## How long does it take the riders to complete trips?
The trips range between 61 seconds (~ 1 mintue) to about 4,000 seconds ( ~ 1 hour) long. However this excludes the outliers which can be as high as 86,000 seconds (~ 24 hours) long. Plotting this on a logarithmic scale, the trip duration has a near normal distribution with a single peak at about 500 seconds ( ~ 8 minutes).

```python
# exclude outliers
outliers = df['duration_sec'] > 4000
df = df.loc[~outliers,:]
# plot distribution on a logarithmic scale
log_binsize = 0.1
bins = 10 ** np.arange(1.7, np.log10(df.duration_sec.max())+log_binsize, log_binsize)

plt.figure(figsize=[8, 5])
df['duration_sec'].hist(bins=bins, grid=False)
plt.xscale('log')
plt.xticks([100, 200, 500, 1000, 2000, 5000], [100, 200, 500, '1k', '2k', '5k'])
plt.xlabel('Trip Duration (seconds)')
plt.ylabel('Frequency')
plt.title('Frequency Distribution of Trip Duration');
```

## What distance is covered during these trips?

For some trips, no distance is covered at all (0km). But for some others, they cover as much as 15 km or more. After excluding outliers and plotting on a logarithmic scale, it can be seen that most trips are between 1 to 2 kilometers wrong.

```python
log_binsize = 0.1
bins = 10 ** np.arange(-1, np.log10(df.distance.max())+log_binsize, log_binsize)
ticks = [0.1, 0.2, 0.5, 1, 2, 5, 10]
labels = [f'{x}' for x in ticks]

plt.figure(figsize=[8, 5])
df['distance'].hist(bins=bins, grid=False)
plt.xscale('log')
plt.xticks(ticks, labels)
plt.xlabel('Distance (kilometres)')
plt.ylabel('Frequency')
plt.title('Frequency Distribution of Distance Covered');
```

## What time of the day do riders prefer to take their trips?

From the graph above, it is seen that people would rather take their trips in the evening or morning. Fewer people take trips at mid day and even fewer people ride at night or early in the morning.

```python
# get a base color so as to reduce noise in plotting
base_color = sb.color_palette()[0]
# plot a barchart showing the frequency distribution of this variable
plt.figure(figsize=[15,7])
sb.countplot(data=df, x='time_of_day', color = base_color)
plt.xlabel('Time of Day')
plt.title('Frequency distribution of the time of day the trip was taken');
```

## How does the trip duration vary with the distance covered?

As the distance increases, there is a slight increase in trip duration. Also, some riders seem to rent bikes but never leave the station. This is explained by the straight line at 0km as seen in the next visual.

```python
plt.figure(figsize=[20,7])
plt.scatter(x='duration_sec', y='distance', data=df, alpha = 1/20)
plt.xlabel('Trip Duration (seconds)')
plt.ylabel('Distance (kilometeres)')
plt.title('Scatter plot showing the correlation between Distance and Trip Duration');
```

## How does the time taken to cover the same distance vary with the time of the day?

The correlation between distance and duration is approximately the same for all times of the day. However, as the day progresses to the evening, more time is taken to cover the same distance. But as night comes, the duration decreases for the same distance.

```python
g = sb.FacetGrid(data = df, col = 'time_of_day', col_wrap= 3, height = 6)
g.map(plt.scatter, 'distance','duration_sec')
g.set_xlabels("Distance Covered (km)")
g.set_ylabels("Trip Duration (seconds)");
```
